```python
import tensorflow as tf
from tensorflow.keras import layers
```

```python
#Transformers Block
class TransformersBlock(layers.Layer):
  def __init__(self,num_heads,embed_dim,ftt_dim,rate=0.1):
    super().__init__()

    self.att = layers.MultiHeadAttention(num_heads=num_heads,key_dim=embed_dim)

    self.ftt = tf.keras.Sequential([
        layers.Dense(ftt_dim,activation='relu'),
        layers.Dense(embed_dim)
    ])

    self.dropout1 = layers.Dropout(rate=rate)
    self.dropout2 = layers.Dropout(rate=rate)

    self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
    self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)

  def call(self,inputs,mask=None):
    #MULTIHEAD
    attention_output = self.att(query=inputs,key=inputs,value=inputs,attention_mask=mask)
    attention_output = self.dropout1(attention_output)

    norm1 = self.layernorm1(inputs + attention_output)

    #MLP
    output = self.ftt(norm1)
    output = self.dropout2(output)
    return self.layernorm2(attention_output + norm1)
```

```python
#Patch Embedding
class PatchEmbedding(layers.Layer):
  def __init__(self,image_size,patch_size,embed_dim):
    super().__init__()
    self.num_patchs = (image_size // patch_size) ** 2
    self.patchEmbed = layers.Dense(embed_dim)

    #Positionembedd
    self.position_embed = layers.Embedding(input_dim=self.num_patchs,output_dim=embed_dim)

  def call(self,input):
    patchEmbedding = self.patchEmbed(input)

    range = tf.range(start=0,limit=self.num_patchs,delta=1)
    positionEmbedding = self.position_embed(range)

    return patchEmbedding + positionEmbedding
```

```python
#VIT
class VisionTransformer(tf.keras.Model):
  def __init__(self,patch_size,image_size,embed_dim,num_heads,ftt_dim,num_classes,num_layers):
    super().__init__()
    self.patch_Embed = PatchEmbedding(image_size=image_size,patch_size=patch_size,embed_dim=embed_dim)

    self.transformer_layer = [
        TransformersBlock(num_heads=num_heads,embed_dim=embed_dim,ftt_dim=ftt_dim)
        for _ in range(num_layers)
    ]

    self.flatten = layers.Flatten()
    self.output = layers.Dense(num_classes,activation='softmax')

  def call(self,images):
    patches = self.extract_Patches(images)
    patch_embedding = self.patch_Embed(patches)

    for transformer_block in self.transformer_layer:
      patch_embedding = transformer_block(patch_embedding)

    flatten = self.flatten(patch_embedding)
    output = self.output(flatten)
    return output

  def extract_Patches(self,images):
    batch_size = tf.shape(images)[0]

    patches = tf.image.extract_patches(
        images= images,
        sizes= [1,patch_size,patch_size,1],
        strides= [1,patch_size,patch_size,1],
        rates= [1,1,1,1],
        padding='VALID'
    )
    patches = tf.reshape(patches,[batch_size,-1,patch_size * patch_size * 3])
    return patches
```
